```python
import random
import math

# Data
data = [
    [1, 2, 1],  # D1 (target)
    [2, 1, 2],  # D2
    [8, 9, 8],  # D3
    [9, 8, 9]   # D4
]

```

```python
# Parameter GA
population_size = 4
crossover_rate = 0.8
mutation_rate = 0.1
elitism_count = 1  

# Target = D1
target = data[0]

```

```python
# Hitung fitness 
def fitness(chromosome):
    total = 0
    i = 0
    while i < len(chromosome):
        selisih = chromosome[i] - target[i]
        total = total + (selisih * selisih)
        i = i + 1
    jarak = total ** 0.5
    if jarak == 0:
        return 1.0
    else:
        return 1 / (1 + jarak)

```

```python
# Crossover satu titik
def crossover(parent1, parent2):
    child = []
    titik = random.randint(1, len(parent1) - 1)
    i = 0
    while i < len(parent1):
        if i < titik:
            child.append(parent1[i])
        else:
            child.append(parent2[i])
        i = i + 1
    return child

```

```python
# Mutasi
def mutate(chromosome):
    hasil = []
    i = 0
    while i < len(chromosome):
        angka = chromosome[i]
        peluang = random.random()
        if peluang < mutation_rate:
            ubah = random.randint(-1, 1)
            angka = angka + ubah
        hasil.append(angka)
        i = i + 1
    return hasil

```

```python
# Inisialisasi populasi
population = data[1:]
generation = 0

# GA Loop dengan kondisi berhenti
while True:
    generation += 1
    print("Generation", generation)

    # Hitung fitness semua individu
    population_fitness = []
    for chrom in population:
        fit = fitness(chrom)
        population_fitness.append((chrom, fit))

    # Urutkan dari yang terbaik
    population_fitness.sort(key=lambda x: x[1], reverse=True)

    # Tampilkan fitness individu
    for i in range(len(population_fitness)):
        chrom, fit = population_fitness[i]
        print(f"  Individu {i+1}: {chrom}, Fitness: {fit:.3f}")

    # Cek apakah solusi terbaik sudah cukup bagus
    if population_fitness[0][1] >= 1.0:
        break

    # Elitism: simpan individu terbaik
    new_population = []
    for i in range(elitism_count):
        new_population.append(population_fitness[i][0])

    # Seleksi dan reproduksi
    while len(new_population) < population_size:
        parent1 = random.choices(population, weights=[fitness(c) for c in population])[0]
        parent2 = random.choices(population, weights=[fitness(c) for c in population])[0]

        if random.random() < crossover_rate:
            child = crossover(parent1, parent2)
        else:
            child = parent1.copy()

        child = mutate(child)
        new_population.append(child)

    # Update populasi
    population = new_population

# Cetak solusi terbaik
best = population_fitness[0][0]
print("\nSolusi terbaik:")
print(f"Kromosom: {best}, Fitness: {fitness(best):.3f}")
```
